```python
def dist_total(df, col):
    '''Take a column in a datafarame and calculate the total, mean and standard deviation of the that column
    '''
    a = df[col].sum()
    m = df[col].mean()
    s = df[col].std()
    return a, m, s
```

```python
total_bill, mena_bill, std_dev = dist_total(df, 'Total Billed Amount')
print("Total Billed :", total_bill.round(2))
print("mean Billed :", mena_bill.round(2))
print("Standard Deviation Billed :", std_dev.round(2))
```

```python
def quad(x,a,b,c):
    return a*x**2 + b*x + c

def quad_solve(a,b,c):
    k = b**2 - 4*a*c
    if k < 0:
        x_1 = (-b + np.sqrt(complex(k)))/(2*a)
        x_2 = (-b - np.sqrt(complex(k)))/(2*a)
    else:
        x_1 = (-b + np.sqrt(k))/(2*a)
        x_2 = (-b - np.sqrt(k))/(2*a)
    return x_1, x_2
```

```python
a,b,c = 2, 3, 1
quad_solve(a,b,c)
```

```python
a,b,c = 2, 3, 4
quad_solve(a,b,c)
```
